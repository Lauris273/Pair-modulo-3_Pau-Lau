```python
def cambiar_comas(cadena):
    try:
        return float(cadena.replace(",", ""))
    except:
        return np.nan
```

```python
cambiar_comas(land_area)
```

```python
df["land_area_modif"] = df["land_area"].apply(cambiar_comas)
```
